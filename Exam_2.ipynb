# ฉันถูกจ้างโดยผู้หญิงคนหนึ่งที่มีเงินมากมาย ฉันไม่ทราบชื่อของเธอ แต่ฉันรู้ว่าเธอสูงประมาณ 5 ฟุต 5 นิ้ว (65 นิ้ว) หรือ 5 ฟุต 7 นิ้ว (67 นิ้ว) เธอมีผมสีแดงและขับรถ Tesla Model S ฉันรู้ว่าเธอไปชมคอนเสิร์ต SQL Symphony 3 ครั้งในเดือนธันวาคม 2017

```python
facebook_event_checkin_df.head()
```

```python
# แปลงคอลัมน์ 'วันที่' ให้เป็นประเภทสตริง
facebook_event_checkin_df['date'] = facebook_event_checkin_df['date'].astype(str)

filtered_facebook_event_checkin_df = facebook_event_checkin_df[
    (facebook_event_checkin_df['event_name'] == 'SQL Symphony Concert') &
    (facebook_event_checkin_df['date'].str.startswith('201712'))
]


# นับจำนวนครั้งที่แต่ละคนเข้าร่วมงาน
event_counts = filtered_facebook_event_checkin_df.groupby('person_id')['event_name'].count()

# กรองหาผู้ที่เข้าร่วมงาน 3 ครั้ง
result_ids = event_counts[event_counts == 3].index.tolist()

# แสดงผลลัพธ์ (person_id ของผู้ที่เข้าร่วมงาน 3 ครั้ง)
filtered_facebook_event_checkin_df[filtered_facebook_event_checkin_df['person_id'].isin(result_ids)]
```
